```python
import numpy

def spinner(probabilityDistribution):
    newProbabilities = []
    for value in probabilityDistribution:
        newProbabilities.append(value / 360)
    index=numpy.random.choice(numpy.arange(0, len(probabilityDistribution)), p=newProbabilities)

    print("Chosen chance:", newProbabilities[index]*100)

probabilityDistribution=[55,85,70,150]
spinner(probabilityDistribution)
```

```python
import numpy
import math
import random
def monteCarlo(R,l,numBall):
    count=0
    for _ in range(numBall):
        x=random.uniform(-l/2,l/2)
        y=random.uniform(-l/2,l/2)
        #pointFromCenter=math.sqrt(x*x+y*y)
        if x**2 + y**2 +1 <= R**2:
            count=count+1
    probability=count/numBall;
    return probability

def mathProb(R,l):
    pipe=math.pi* (R*R)
    rectArea=l*l;
    probability=pipe/rectArea
    return probability
        
R=30
l=100

numBall=500000

testedProb=monteCarlo(R,l,numBall)
mathProbability=mathProb(R,l)

print("MonteCarlo probability:",testedProb)
print("Math probability:", mathProbability)
```
